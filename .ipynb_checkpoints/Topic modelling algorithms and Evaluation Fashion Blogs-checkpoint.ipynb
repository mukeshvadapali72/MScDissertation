### Reading Libraries

```python
import pandas as pd
import pickle
import numpy as np
from numpy import array
import logging, gensim, bz2
import matplotlib.pyplot as plt 
from gensim.models import CoherenceModel, LdaModel, LsiModel, HdpModel

logging.basicConfig(format='%(asctime)s : %(levelname)s : %(message)s', level=logging.INFO)
```

```python
webblogsdocs =array(data_clean['fashiontext'])
```

### Preprocessing for Topic Model

```python
from nltk.tokenize import RegexpTokenizer

def docs_preprocessor(docs):
    tokenizer = RegexpTokenizer(r'\w+')
    for idx in range(len(docs)):
        docs[idx] = docs[idx].lower()  # Convert to lowercase.
        docs[idx] = tokenizer.tokenize(docs[idx])  # Split into words.

    # Remove numbers, but not words that contain numbers.
    docs = [[token for token in doc if not token.isdigit()] for doc in docs]
    
    # Remove words that are only one character.
    docs = [[token for token in doc if len(token) > 3] for doc in docs] #considering length of tokens more than 3
    
  
    return docs
```

### Apply the function

```python
webblogsdocs = docs_preprocessor(webblogsdocs)
```

### Creating Bigrams and Trigrams

```python
#Create Biagram & Trigram Models 
from gensim.models import Phrases
# Add bigrams and trigramsa to docs,minimum count 10 means only that appear 10 times or more.
bigram = Phrases(webblogsdocs, min_count=10)
trigram = Phrases(bigram[webblogsdocs])

for idx in range(len(webblogsdocs)):
    for token in bigram[webblogsdocs[idx]]:
        if '_' in token:
            # Token is a bigram, add to document.
            webblogsdocs[idx].append(token)
    for token in trigram[webblogsdocs[idx]]:
        if '_' in token:
            # Token is a bigram, add to document.
            webblogsdocs[idx].append(token)
```

#### In order to work on text documents, Gensim Topic models requires the words (aka tokens) be converted to unique ids. In order to achieve that, Gensim lets you create a Dictionary object that maps each word to a unique id.

```python
# Create a dictionary representation of the documents.
from gensim.corpora.dictionary import Dictionary
dictionary = Dictionary(webblogsdocs)
```

The dictionary object is typically used to create a ‘bag of words’ Corpus. It is this Dictionary and the bag-of-words (Corpus) that are used as inputs to topic modeling and other models that Gensim specializes in

#### ‘Corpus’ is typically a ‘collection of documents as a bag of words’. That is, for each document, a corpus contains each word’s id and its frequency count in that document. As a result, information of the order of words is lost.

```python
#Create dictionary and corpus required for Topic Modeling
corpus = [dictionary.doc2bow(doc) for doc in webblogsdocs]
print('Number of unique tokens: %d' % len(dictionary))
print('Number of documents: %d' % len(corpus))
```

#### Code to show output of topics with maximum 15 words in each topic in dataframe format

```python
def get_topics(model, num_topics):
    word_dict = {};
    for i in range(num_topics):
        words = model.show_topic(i, topn = 15);
        word_dict['Topic ' + '{:02d}'.format(i+1)] = [i[0] for i in words];
    return pd.DataFrame(word_dict);
```

### Using POSTagging considering only Noun and Adjective

```python
from nltk import word_tokenize, pos_tag
def nouns_adj(text):
    '''Given a string of text, tokenize the text and pull out only the nouns and adjectives.'''
    is_noun_adj = lambda pos: pos[:2] == 'NN' or pos[:2] == 'JJ'
    tokenized = word_tokenize(text)
    nouns_adj = [word for (word, pos) in pos_tag(tokenized) if is_noun_adj(pos)] 
    return ' '.join(nouns_adj)
```

```python
data_nouns_adj = pd.DataFrame(data_clean.fashiontext.apply(nouns_adj))
data_nouns_adj
```

## LSI Model

#### Computing coherence measure for random number of topics

```python
def compute_coherence_values(dictionary, corpus, texts, limit, start=2, step=3):
    
    coherence_values = []
    model_list = []
    for num_topics in range(start, limit, step):
        
        model = LsiModel(corpus=corpus, num_topics=num_topics, id2word=dictionary)
        model_list.append(model)
        coherencemodel = CoherenceModel(model=model, texts=texts, dictionary=dictionary, coherence='c_v')
        coherence_values.append(coherencemodel.get_coherence())

    return model_list, coherence_values
```

```python
model_list, coherence_values = compute_coherence_values(dictionary=dictionary, corpus=corpus3, texts=webblogsdocs2, start=2, limit=25, step=2)
# Show graph
import matplotlib.pyplot as plt
limit=25; start=2; step=2;
x = range(start, limit, step)
plt.plot(x, coherence_values)
plt.xlabel("Num Topics")
plt.ylabel("Coherence score")
plt.legend(("coherence_values"), loc='best')
plt.show()
```

```python
#new
for m, cv in zip(x, coherence_values):
    print('Num Topics =', m, ' has Coherence Value of LSIModel', round(cv, 4))
```

### LDA Model

#### Computing coherence measure for random number of topics

```python
def compute_coherence_values(dictionary, corpus, texts, limit, start=2,step=2):
    
    coherence_values = []
    model_list = []
    for num_topics in range(start, limit):
        model=LdaModel(corpus=corpus, id2word=dictionary, num_topics=num_topics)
        model_list.append(model)
        coherencemodel = CoherenceModel(model=model, texts=texts, dictionary=dictionary, coherence='c_v')
        coherence_values.append(coherencemodel.get_coherence())

    return model_list, coherence_values
```

```python
model_list, coherence_values = compute_coherence_values(dictionary=dictionary, corpus=corpus3, texts=webblogsdocs2, start=2, limit=25,step=2)
# Show graph
import matplotlib.pyplot as plt
limit=25; start=2;
x = range(start, limit)
plt.plot(x, coherence_values)
plt.xlabel("Num Topics")
plt.ylabel("Coherence score")
plt.legend(("coherence_values"), loc='best')
plt.show()
```

```python
#Print the coherence scores
for m, cv in zip(x, coherence_values):
    print('Num Topics =', m, ' has Coherence Value of', round(cv, 4))
```

## Evaluating topic models

```python
lsitopics = [[word for word, prob in topic] for topicid, topic in lsimodel3.show_topics(formatted=False)]

hdptopics = [[word for word, prob in topic] for topicid, topic in hdpmodel3.show_topics(formatted=False)]

ldatopics = [[word for word, prob in topic] for topicid, topic in ldamodel3.show_topics(formatted=False)]
```

### UMass measures to compare a word only to the preceding and succeeding words respectively, so need ordered word set.It uses as pairwise score function which is the empirical conditional log-probability with smoothing count to avoid calculating the logarithm of zero.

```python
lsi_coherenceumass = CoherenceModel(topics=lsitopics, texts=webblogsdocs2, dictionary=dictionary, window_size=10,coherence='u_mass').get_coherence()

hdp_coherenceumass = CoherenceModel(topics=hdptopics, texts=webblogsdocs2, dictionary=dictionary, window_size=10,coherence='u_mass').get_coherence()

lda_coherenceumass = CoherenceModel(topics=ldatopics, texts=webblogsdocs2, dictionary=dictionary, window_size=10,coherence='u_mass').get_coherence()
```

```python
print('umass lsi coherence measure',lsi_coherenceumass)
print('umass hdp coherence measure',hdp_coherenceumass )
print('umass lda coherence measure',lda_coherenceumass )
```

```python
def evaluate_bar_graph(coherences, indices):
    """
    Function to plot bar graph.
    
    coherences: list of coherence values
    indices: Indices to be used to mark bars. Length of this and coherences should be equal.
    """
    assert len(coherences) == len(indices)
    n = len(coherences)
    x = np.arange(n)
    plt.bar(x, coherences, width=0.2, tick_label=indices, align='center')
    plt.xlabel('Models')
    plt.ylabel('Coherence Value')
    plt.gca().invert_yaxis()
```

```python
evaluate_bar_graph([lsi_coherenceumass, hdp_coherenceumass, lda_coherenceumass],
                   ['LSI', 'HDP', 'LDA'])
```
