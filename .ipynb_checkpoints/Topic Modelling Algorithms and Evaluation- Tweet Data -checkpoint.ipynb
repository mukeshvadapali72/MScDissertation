```python
import pandas as pd
import numpy as np
from numpy import array
import logging, gensim, bz2
import matplotlib.pyplot as plt
from nltk.stem.wordnet import WordNetLemmatizer
from nltk.tokenize import RegexpTokenizer
from gensim.models import Phrases
from gensim.corpora.dictionary import Dictionary
from gensim.models import CoherenceModel, LdaModel, LsiModel, HdpModel

#logging.basicConfig(format='%(asctime)s : %(levelname)s : %(message)s', level=logging.INFO)
```

### Reading data

```python
fashion_data = pd.read_pickle('data_cleanfashiontweet.pkl')
```

### Convert to array

```python
docs =array(fashion_data['text'])
```

### Preprocessing of data 

```python
def docs_preprocessor(docs):
    tokenizer = RegexpTokenizer(r'\w+')
    for idx in range(len(docs)):
        docs[idx] = docs[idx].lower()  # Convert to lowercase.
        docs[idx] = tokenizer.tokenize(docs[idx])  # Split into words.

    # Remove numbers.
    docs = [[token for token in doc if not token.isdigit()] for doc in docs]
    
    # Remove words with length less than 3.
    docs = [[token for token in doc if len(token) > 3] for doc in docs] #considering length of tokens more than 3
    
    # Lemmatize all words in documents.
    lemmatizer = WordNetLemmatizer()
    docs = [[lemmatizer.lemmatize(token) for token in doc] for doc in docs]
  
    return docs
```

```python
docs = docs_preprocessor(docs)
```

### Create Biagram & Trigram Models 

```python
# Add bigrams and trigrams to docs,minimum count 25 means only that appear 10 times or more.
bigram = Phrases(docs, min_count=25)
trigram = Phrases(bigram[docs])

for idx in range(len(docs)):
    for token in bigram[docs[idx]]:
        if '_' in token:
            # Token is a bigram, add to document.
            docs[idx].append(token)
    for token in trigram[docs[idx]]:
        if '_' in token:
            # Token is a bigram, add to document.
            docs[idx].append(token)
```

## Data Transformation

```python
dictionary = Dictionary(docs)
```

```python
corpus = [dictionary.doc2bow(doc) for doc in docs]
print('Number of unique tokens: %d' % len(dictionary))
print('Number of documents: %d' % len(corpus))
```

## Topic Modelling

```python
def get_topics(model, num_topics):
    word_dict = {};
    for i in range(num_topics):
        words = model.show_topic(i, topn = 20);
        word_dict['Topic ' + '{:02d}'.format(i+1)] = [i[0] for i in words];
    return pd.DataFrame(word_dict);
```

### LDA Topic Model

### Noun and Adjective

```python
from nltk import word_tokenize, pos_tag
def nouns_adj(text):
    '''Given a string of text, tokenize the text and pull out only the nouns and adjectives.'''
    is_noun_adj = lambda pos: pos[:2] == 'NN' or pos[:2] == 'JJ'
    tokenized = word_tokenize(text)
    nouns_adj = [word for (word, pos) in pos_tag(tokenized) if is_noun_adj(pos)] 
    return ' '.join(nouns_adj)
```

```python
data_nouns_adj = pd.DataFrame(fashion_data['text'].apply(nouns_adj))
data_nouns_adj
```

```python
fashiondocs2 =array(data_nouns_adj['text'])
```

```python
fashiondocs2 = docs_preprocessor(fashiondocs2)
```

```python
bigram = Phrases(fashiondocs2, min_count=10)
trigram = Phrases(bigram[fashiondocs2])

for idx in range(len(fashiondocs2)):
    for token in bigram[fashiondocs2[idx]]:
        if '_' in token:
            # Token is a bigram, add to document.
            fashiondocs2[idx].append(token)
    for token in trigram[fashiondocs2[idx]]:
        if '_' in token:
            # Token is a bigram, add to document.
            fashiondocs2[idx].append(token)
```

```python
dictionary = Dictionary(fashiondocs2)
```

```python
corpus2 = [dictionary.doc2bow(doc) for doc in fashiondocs2]

```

### LDA MODEL

```python
def compute_coherence_values(dictionary, corpus, texts, limit, start=2, step=2):
    
    coherence_values = []
    model_list = []
    for num_topics in range(start, limit, step):
        model=LdaModel(corpus=corpus, id2word=dictionary, num_topics=num_topics)
        model_list.append(model)
        coherencemodel = CoherenceModel(model=model, texts=texts, dictionary=dictionary, coherence='c_v')
        coherence_values.append(coherencemodel.get_coherence())

    return model_list, coherence_values
```

```python
model_list, coherence_values = compute_coherence_values(dictionary=dictionary, corpus=corpus2, texts=fashiondocs2, start=2, limit=25, step=2)
# Show graph
limit=25; start=2; step=2;
x = range(start, limit, step)
plt.plot(x, coherence_values)
plt.xlabel("Num Topics")
plt.ylabel("Coherence score")
plt.legend(("coherence_values"), loc='best')
plt.show()
plt.savefig('lda coherence measure twitter')
```

#### Print the coherence scores

```python
for m, cv in zip(x, coherence_values):
    print('Num Topics =', m, ' has Coherence Value of', round(cv, 4))
```

### LSI Model

```python
def compute_coherence_values(dictionary, corpus, texts, limit, start=2, step=3):
    
    coherence_values = []
    model_list = []
    for num_topics in range(start, limit, step):
        
        model = LsiModel(corpus=corpus, num_topics=num_topics, id2word=dictionary)
        model_list.append(model)
        coherencemodel = CoherenceModel(model=model, texts=texts, dictionary=dictionary, coherence='c_v')
        coherence_values.append(coherencemodel.get_coherence())

    return model_list, coherence_values
```

```python
model_list, coherence_values = compute_coherence_values(dictionary=dictionary, corpus=corpus2, texts=fashiondocs2, start=2, limit=25, step=2)
# Show graph
import matplotlib.pyplot as plt
limit=25; start=2; step=2;
x = range(start, limit, step)
plt.plot(x, coherence_values)
plt.xlabel("Num Topics")
plt.ylabel("Coherence score")
plt.legend(("coherence_values"), loc='best')
plt.show()
```

```python
for m, cv in zip(x, coherence_values):
    print('Num Topics =', m, ' has Coherence Value of LSIModel', round(cv, 4))
```

## Evaluation of topic model

```python
lsitopics = [[word for word, prob in topic] for topicid, topic in lsimodel2.show_topics(formatted=False)]

hdptopics = [[word for word, prob in topic] for topicid, topic in hdpmodel2.show_topics(formatted=False)]

ldatopics = [[word for word, prob in topic] for topicid, topic in ldamodel2.show_topics(formatted=False)]
```

```python
def evaluate_bar_graph(coherences, indices):
    """
    Function to plot bar graph.
    
    coherences: list of coherence values
    indices: Indices to be used to mark bars. Length of this and coherences should be equal.
    """
    assert len(coherences) == len(indices)
    n = len(coherences)
    x = np.arange(n)
    plt.bar(x, coherences, width=0.2, tick_label=indices, align='center')
    plt.xlabel('Models')
    plt.ylabel('Coherence Value')
    plt.gca().invert_yaxis()
```

```python
lsi_coherenceumass = CoherenceModel(topics=lsitopics, texts=fashiondocs2, dictionary=dictionary, window_size=10,coherence='u_mass').get_coherence()

hdp_coherenceumass = CoherenceModel(topics=hdptopics, texts=fashiondocs2, dictionary=dictionary, window_size=10,coherence='u_mass').get_coherence()

lda_coherenceumass = CoherenceModel(topics=ldatopics, texts=fashiondocs2, dictionary=dictionary, window_size=10,coherence='u_mass').get_coherence()
```

```python
import matplotlib.pyplot as plt
```

```python
evaluate_bar_graph([lsi_coherenceumass, hdp_coherenceumass, lda_coherenceumass],
                   ['LSI', 'HDP', 'LDA'])
```
